## Pré-processamento

Aqui ficará o pré-processamento dos datasets.

```python
def tratar_linha(linha):
    if (linha[-2:] == ";\n"):
        linha = linha[:-2]
    return linha

def preprocessar_turistas(nomeCsv):
    with open("Datasets/" + nomeCsv) as original:
        with open("Datasets/Pre-processados/" + nomeCsv, 'w+') as novo:
            variaveis = original.readline()\
                                .lower()\
                                .replace("ordem", "cod")\
                                .replace("mes", "mês")
            tratar_linha(variaveis)
            print(variaveis)
            novo.write(variaveis)
            for line in original:
                line = tratar_linha(line)
                print(line)
                novo.write(line + '\n')

preprocessar_turistas("chegadas_2014.csv")
```

```python
def leitura_turistas(csvURL):
    saida = pd.read_csv(csvURL, sep=';', encoding='latin1',\
                        usecols=["Continente", "País", "UF", "cod mes", "Chegadas"]\
                       )
    saida['Continente'] = saida['Continente'].astype('category')
    saida['País'] = saida['País'].astype('category')
    saida['UF'] = saida['UF'].astype('category')
    return pd.DataFrame(saida.groupby(["Continente", "País", "UF", "cod mes"]).Chegadas.sum())
```
