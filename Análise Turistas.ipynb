## Pré-processamento

Aqui ficarão as funções de leitura e pré-processamento dos datasets.

```python
#Dicionario para converter o texto do mês em número
mes_string_int = {
    'Janeiro': 1,
    'Fevereiro': 2,
    'Março': 3,
    'Abril': 4,
    'Maio': 5,
    'Junho': 6,
    'Julho': 7,
    'Agosto': 8,
    'Setembro': 9,
    'Outubro': 10,
    'Novembro': 11,
    'Dezembro': 12
}

def remove_dots(x):
    '''Remove os pontos de uma string, usado para remover os pontos que vêm nos números de chegadas.'''
    return x.replace('.', '')

def leitura_turistas(ano):
    '''Lê os CSVs dos turistas.'''
    saida = pd.read_csv('Datasets/chegadas_' + str(ano) + '.csv', sep=';', encoding='latin1', skiprows=1,\
                        dtype={'Continente': 'category', 'País': 'category',\
                               'UF': 'category', 'Mês': 'category', 'Chegadas': object},\
                        names=['Continente', 'País', 'UF', 'Mês', 'Chegadas'])
    saida['Mês'] = saida['Mês'].apply(lambda x: mes_string_int[x]).astype(np.uint8)
    saida['Chegadas'] = saida['Chegadas'].apply(remove_dots).astype(np.uint32)
    return pd.DataFrame(saida)
```

## Carregamento

Aqui os datasets serão carregados

```python
chegadas_2014 = leitura_turistas(2014)
chegadas_2015 = leitura_turistas(2015)
chegadas_2016 = leitura_turistas(2016)
chegadas_2017 = leitura_turistas(2017)
chegadas_2014
```

## Análise exploratória

Observação dos dados com a intenção de achar coisas interessantes para trabalhar.

### Turistas por mês

Observando os gráficos das chegadas por mês, pode-se perceber que elas tendem a aumentar no verão e diminuir no inverno. Porém, existe uma exceção. No ano de 2014 há uma grande quantidade de turistas durante os meses de junho e julho. Esses foram os meses durante os quais estava ocorrendo a Copa do Mundo que foi sediada no Brasil.

```python
%%opts Bars [width=400, height=250, tools=['hover']] (color='#ff0000')
bars_2014_2017 = hv.Layout(
    hv.Bars(chegadas_2014, kdims='Mês', vdims='Chegadas', group='2014').aggregate('Mês', function=np.sum)\
     + hv.Bars(chegadas_2015, kdims='Mês', vdims='Chegadas', group='2015').aggregate('Mês', function=np.sum)\
     + hv.Bars(chegadas_2016, kdims='Mês', vdims='Chegadas', group='2016').aggregate('Mês', function=np.sum)\
     + hv.Bars(chegadas_2017, kdims='Mês', vdims='Chegadas', group='2017').aggregate('Mês', function=np.sum)).cols(2)
display(bars_2014_2017)
```

### 2014

```python
%%opts HeatMap [colorbar=True width=480 height=300 tools=['hover'] toolbar='above']
def heatmap_mes_UF(df, group):
    '''Função que produz o HeatMap do ano do dataFrame dado.'''
    dataset = hv.Dataset(df, vdims='Chegadas')
    heatmap = hv.HeatMap(dataset.aggregate(['Mês', 'UF'], np.sum), group=group)
    return heatmap

display(hv.Layout(heatmap_mes_UF(chegadas_2014, '2014')
                + heatmap_mes_UF(chegadas_2015, '2015')
                + heatmap_mes_UF(chegadas_2016, '2016')
                + heatmap_mes_UF(chegadas_2017, '2017')).cols(2))
```
