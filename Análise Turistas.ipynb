## Imports

Aqui devem ficar os imports, assim só os executo uma vez.

```python
import pandas as pd
import numpy as np
```

```python
# Leitura do csv
chegadas_2016 = pd.read_csv("Datasets/chegadas_2016.csv", sep=';', encoding='latin1',\
                            usecols=["Continente", "País", "UF", "cod mes", "Chegadas"]\
                           )

# Mudança de tipo das variaveis
chegadas_2016['Continente'] = chegadas_2016['Continente'].astype('category')
chegadas_2016['País'] = chegadas_2016['País'].astype('category')
chegadas_2016['UF'] = chegadas_2016['UF'].astype('category')

# União de entradas repetidas (que existem porque não usei todas as variaveis)
chegadas_2016 = pd.DataFrame(chegadas_2016.groupby(["Continente", "País", "UF", "cod mes"]).Chegadas.sum())
chegadas_2016
```

```python
chegadas_2016['Chegadas'].sum()
```

```python
pd.DataFrame(chegadas_2016.groupby(["Continente"]).Chegadas.sum()).sort_values('Chegadas', ascending=False)
```

```python
pd.DataFrame(chegadas_2016.groupby(["País"]).Chegadas.sum()).sort_values('Chegadas', ascending=False)
```

```python
pd.DataFrame(chegadas_2016.groupby(["UF"]).Chegadas.sum()).sort_values('Chegadas', ascending=False)
```
