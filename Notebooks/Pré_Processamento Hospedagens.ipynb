```python
import pandas as pd
import numpy as np
```

```python
##chegadasTuristas = pd.read_csv('Datasets/chegadas_2016.csv' erro

##Carregando dados
hospedagens_trimestre01 = pd.read_csv('http://turismo.gov.br/dadosabertos/cadasturpj/MeiodeHospedagem201601TrimestreCadasturPJ.csv',encoding='latin-1',sep=';')
hospedagens_trimestre02 = pd.read_csv('http://turismo.gov.br/dadosabertos/cadasturpj/MeiodeHospedagem201602TrimestreCadasturPJ.csv',encoding='latin-1',sep=';')
hospedagens_trimestre03 = pd.read_csv('http://www.turismo.gov.br/dadosabertos/cadasturpj/MeiodeHospedagem201603TrimestreCadasturPJ.csv',encoding='latin-1',sep=';')
hospedagens_trimestre04 = pd.read_csv('http://www.turismo.gov.br/dadosabertos/cadasturpj/MeiodeHospedagem201604TrimestreCadasturPJ.csv',encoding='latin-1',sep=';')

```

```python
## Colunas Importantes
hospedagens_trimestre01.columns
```

```python
##Escolhendo colunas
dados_hospedagens_trimestre01 = pd.DataFrame(hospedagens_trimestre01,columns = ['UF', 'SITUAÇÃO', 'LINGUAS']) 

## Hospedagens que estão funcionando
dados_hospedagens_trimestre01 = dados_hospedagens_trimestre01[dados_hospedagens_trimestre01.SITUAÇÃO.str.contains('Em Operação')]
```

```python
## Eliminando dados inválidos
dados_hospedagens_trimestre01 = dados_hospedagens_trimestre01.dropna()
```

```python
dados_hospedagens_trimestre01.head()
```

```python
from nltk.tokenize import sent_tokenize , word_tokenize
allIdiomas = []
for valor in dados_hospedagens_trimestre01.LINGUAS:
    valor_tokenize = word_tokenize(valor)
    for palavra in valor_tokenize:
        if (len(palavra) > 4) and (palavra != 'Nenhuma') and (palavra != 'Opções'):
            allIdiomas.append(palavra)

linguas = []
for idioma in allIdiomas:
    if idioma not in linguas:
        linguas.append(idioma)

# Idiomas oferecidos pelas hospedagens
print(linguas)
```

```python
# Adcionando Novas colunas
for lingua in linguas:
    dados_hospedagens_trimestre01[lingua] = 0
```

```python
print(dados_hospedagens_trimestre01.dtypes)
dados_hospedagens_trimestre01.head()
```

```python
# Adicionando valores as coluncas de idiomas oferecidos
for lingua in linguas:
    dados_hospedagens_trimestre01.loc[dados_hospedagens_trimestre01['LINGUAS'].str.contains(lingua),lingua] = 1
```
